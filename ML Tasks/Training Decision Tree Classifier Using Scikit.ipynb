## Training Decision Tree Classifier Using Scikit 

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
```

```python
address = '../data/iris.csv'

dataset = pd.read_csv(address)
dataset.head()
```

```python
dataset.Species.unique()
```

### Separating features and labels

```python
X = dataset.iloc[:, 1:5]
X
```

```python
y = dataset.iloc[:,5]
y
```

### Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0) # random_state=0 to get the same result every time
```

### Training Decision Tree Classifier

```python
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
```

```python
y_predict = clf.predict(X_test)
y_predict
```

### Evaluation metric

```python
accuracy = metrics.accuracy_score(y_test, y_predict)
print(f'Accuracy: {accuracy}')
```
